**MultiAgent**

Reward

```python
# S1: AV company
reward_data_S1 = np.load('/content/Rt_8724718_episodes_S1.npy', allow_pickle=True)
# S2: Human-drivers
reward_data_S2 = np.load('/content/Rt_8724718_episodes_S2.npy', allow_pickle=True)

plt.figure(figsize=(6,8), dpi=80)
ax1 = plt.subplot(211)
ax1.plot(np.arange(len(reward_data_S1))*200, reward_data_S1)
plt.tick_params('x', labelbottom=False)
ax1.set_title("Reward Averaged Every 200 Episodes for S1")
plt.ylabel('Average Reward')

ax2 = plt.subplot(212, sharex=ax1)
ax2.plot(np.arange(len(reward_data_S2))*200, reward_data_S2)
ax2.set_title("Reward Averaged Every 200 Episodes for S2")
plt.ylabel('Average Reward')

```
